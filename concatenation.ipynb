```python
import numpy as np
import pandas as pd
```

```python
data_one = {'A': ['A0','A1','A2','A3'], 'B': ['B0','B1','B2','B3']}
data_two = {'C': ['C0','C1','C2','C3'], 'D': ['D0','D1','D2','D3']}
```

```python
one = pd.DataFrame(data_one)
```

```python
two = pd.DataFrame(data_two)
```

```python
one 
```

```python
two
```

```python
pd.concat([one,two],axis=1)
```

```python
pd.concat([two,one],axis=1)
```

```python
pd.concat([one,two],axis=0)
```
